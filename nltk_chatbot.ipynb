```python
import random
import nltk
from nltk.stem.lancaster import LancasterStemmer

# Download nltk packages (only first time)
try:
    nltk.data.find("tokenizers/punkt")
except LookupError:
    nltk.download("punkt")

stemmer = LancasterStemmer()

# ---- INTENTS DATA ----
intents = {
    "greeting": {
        "patterns": ["hello", "hi", "hey", "good morning", "good afternoon"],
        "responses": ["Hello!", "Hi there!", "Hey!", "Good to see you!"]
    },
    "goodbye": {
        "patterns": ["bye", "goodbye", "see you"],
        "responses": ["Goodbye!", "See you later!", "Take care!"]
    },
    "thanks": {
        "patterns": ["thanks", "thank you", "thanks a lot"],
        "responses": ["You're welcome!", "Happy to help!", "Anytime!"]
    },
    "about": {
        "patterns": ["who are you", "what are you", "tell me about you"],
        "responses": ["I'm a Python chatbot!", "I'm your virtual assistant!"]
    },
    "help": {
        "patterns": ["help me", "i need help", "can you help", "assist me"],
        "responses": ["Sure! What do you need help with?", "I'm here to assist you!"]
    }
}

# ---- FUNCTIONS ----
def tokenize(sentence):
    return sentence.lower().split()

def stem_words(words):
    return [stemmer.stem(w) for w in words]

def match_intent(message):
    words = stem_words(tokenize(message))
    for intent, data in intents.items():
        for pattern in data["patterns"]:
            pattern_words = stem_words(tokenize(pattern))
            if any(w in pattern_words for w in words):
                return intent
    return None

# ---- CHAT LOOP ----
def chatbot():
    print("Chatbot: Hello! Type 'quit' to exit.\n")

    while True:
        user_input = input("You: ")

        if user_input.lower() == "quit":
            print("Chatbot: Goodbye!")
            break

        intent = match_intent(user_input)

        if intent:
            response = random.choice(intents[intent]["responses"])
            print("Chatbot:", response)
        else:
            print("Chatbot: I didn't understand that. Try again!")


# ---- START CHATBOT ----
if __name__ == "__main__":
    chatbot()
```
